```python
import os
import sqlite3
from pathlib import Path

import pandas as pd

# In notebooks, __file__ is not defined. Use the current working directory.
# If you run the notebook from CodeCity/notebooks, project root is one level up.
CWD = Path.cwd()
BASE_DIR = CWD.parent if CWD.name == "notebooks" else CWD
DB_PATH = BASE_DIR / "codecity.db"

print("Using DB:", DB_PATH)

conn = sqlite3.connect(str(DB_PATH))

query = """
SELECT
  snapshot_id,
  path,
  name,
  size,
  complexity,
  width,
  depth,
  height,
  extension,
  is_test_file,
  area,
  aspect_ratio,
  risk_score,
  anomaly_score
FROM files
"""

files_df = pd.read_sql_query(query, conn)
conn.close()

display(files_df.head(10))
print("Rows:", len(files_df))
display(files_df.describe(include="all"))
```
