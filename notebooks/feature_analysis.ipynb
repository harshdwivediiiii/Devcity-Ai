# Code Feature Analysis

This notebook explores the distribution of code features such as logic complexity, file size, depth, and churn, built via `src/ml/feature_builder.py`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example of loading mock records extracted from git logic
from src.ml.feature_builder import build_features

records = [
    {"size": 120, "complexity": 12, "depth": 2, "extension": ".py", "churn": 5},
    {"size": 500, "complexity": 45, "depth": 4, "extension": ".ts", "churn": 20},
    {"size": 40, "complexity": 2, "depth": 1, "extension": ".md", "churn": 1}
]

features = [build_features(r) for r in records]
df = pd.DataFrame(features)
df.head()
```

```python
sns.pairplot(df)
plt.suptitle('Pairplot of Code Entity Features', y=1.02)
plt.show()
```
