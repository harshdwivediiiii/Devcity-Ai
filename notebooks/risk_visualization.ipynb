# Risk Scoring Visualization

This notebook acts as an analysis plane over the risk assignments generated by the ML pipeline in `src/ml/risk_model.py`.
We visualize how factors such as code size and churn mathematically translate into overall risk scores.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from src.ml.risk_model import compute_risk_score

# Generate syntheic distribution for risk evaluation
np.random.seed(42)
n_samples = 100
mock_features = pd.DataFrame({
    "size": np.random.randint(10, 15000, n_samples),
    "complexity": np.random.randint(1, 60, n_samples),
    "depth": np.random.randint(1, 10, n_samples),
    "type_importance": np.random.uniform(0.1, 1.0, n_samples),
    "churn": np.random.randint(0, 30, n_samples)
})

mock_features['risk_score'] = mock_features.apply(lambda row: compute_risk_score(row.to_dict()), axis=1)
mock_features.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(mock_features['risk_score'], bins=20, kde=True)
plt.title('Distribution of ML-Generated Code Risk Scores')
plt.xlabel('Risk Score (0.0 - 1.0)')
plt.ylabel('File Count')
plt.show()
```
