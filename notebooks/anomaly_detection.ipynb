# Anomaly Detection using Z-Scores

This notebook explores the execution of `src/ml/anomaly_detector.py`. We detect outliers in complexity and size within a codebase.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from src.ml.anomaly_detector import detect_anomalies

# Build normal cluster, and two outliers
complexities = list(np.random.normal(10, 2, 98))
complexities.extend([45, 52]) # The anomalies

features = [{"complexity": c} for c in complexities]
anomalies = detect_anomalies(features)
df = pd.DataFrame({"complexity": complexities, "is_anomaly": anomalies})
df.sort_values(by='complexity', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(range(len(df)), df['complexity'], c=df['is_anomaly'], cmap='coolwarm', alpha=0.8)
plt.title('Code Complexity Over Files (Red indicates Z-score anomaly flagged)')
plt.xlabel('File Index')
plt.ylabel('Syntactic Complexity')
plt.show()
```
